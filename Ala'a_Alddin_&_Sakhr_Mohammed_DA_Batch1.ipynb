#Analyzing the most common columns in the dataset

```python
 # function to plot pie and count
def plot_pie_count(colunas):
    Fig, ax = plt.subplots(1,2, figsize=(14,7))
    
    plt.suptitle(f'{colunas}', fontsize=16, fontweight='bold')
    store[colunas].value_counts().plot.pie(autopct='%.2f%%', textprops={'size':14},wedgeprops={'width':0.6, 'edgecolor':'white'},ax=ax[0])
    ax[0].axis('off')
    sns.countplot(x=store[colunas],ax=ax[1],alpha=0.5)
    ax[1].tick_params(axis='x', labelsize=14)
    for i in ax[1].containers:
        ax[1].bar_label(i, label_type='center', rotation=90, fontsize=16)
    plt.xlabel(None)
    plt.tight_layout()
    plt.show()
```

```python
columns= ['ship_mode','segment', 'market', 'region','category', 'sub_category', 'order_priority', 'year']
for i in columns:
    plot_pie_count(i)
    print()
    
```

# calculate net profit

```python
#define profit and loss
store_profit=store[store['profit'] > 0]
store_loss=store[store['profit'] < 0]
```

```python
#Shape of the profit Dataframe
store_profit.shape
```

```python
#Shape of the loss Dataframe
store_loss.shape
```

```python
#calculating total profit
Total_Profit = store_profit['profit'].sum()
Total_Profit
```

```python
#calculating total loss
Total_loss=np.negative(store_loss['profit'].sum())
Total_loss
```

```python
Net_Profit = Total_Profit - Total_loss
Net_Profit
```

```python
#create a dataFrame o visualize the bervious steps
Df_Net_Profit=pd.DataFrame({'Total_Profit': {0: 2389392},
                            'Total_loss':{0: 920357},
                            ''         '':{0: '='},
                            'Net_Profit':{0: 1469035}})
Df_Net_Profit

```

```python

Df_Net_Profit.plot(kind='bar')
plt.xlabel("profits")
plt.ylabel("In millions")
```
